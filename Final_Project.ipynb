```python
# importing the data and the libraries needed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('education_career_success.csv')
# .head() to get a glance at the data
df.head()
```

```python
df.info()
```

```python
df.describe()
```

I used .info() and .describe() to check for outliers, missing values(null), duplicates, and to get a better understanding of the data. None of these were found so the only thing I am going to do is remove the Entrepreneurship column.

```python
#removing the Entrepreneurship column because it isn't nessesary for my analysis.
df.drop(columns=['Entrepreneurship'], inplace=True)
df.head()
```

```python
# Data Manipulation: Sort by age in ascending order
# This will help me see the range of ages I am getting my data from
# It makes sense why none of them has started there own business because they are still pretty young.
df_sorted_by_age = df.sort_values(by='Age')
df_sorted_by_age
```

```python
# This displayed the count of each columns data
df_count = df.count()
df_count
```

```python
# I wanted to see what the percentage of Males vs Females was, they are 50% each.
df['Gender'].value_counts()
```

```python
# Data Filtering to see if the people with the highest job offers also had high numbers in the rest of the columns, they did.
job_offers_most = df[df['Job_Offers'] == 5]
job_offers_most
```

```python
# Data Filtering to see how many people had a low GPA in high school
low_GPA = df[df['High_School_GPA'] <= 3.4]
low_GPA
```

```python
# Counting the amount of people in each different job level, the least amount were in Senior which makes sense because when we looked at the ege they were all still fairly young.
df['Current_Job_Level'].value_counts()
```

```python
# Discriptive Statistics on the starting salary
print('Total rows', df['Starting_Salary'].shape)
print('Count', df['Starting_Salary'].count())
print('Count Unique', df['Starting_Salary'].nunique())
print('Sum', df['Starting_Salary'].sum())
print('Average', df['Starting_Salary'].mean())
print('Median', df['Starting_Salary'].median())
print('25th percentile', df['Starting_Salary'].quantile(0.25))
print('75th percentile', df['Starting_Salary'].quantile(0.75))
print('Standard Deviation', df['Starting_Salary'].std())
print('Minimum', df['Starting_Salary'].min())
print('Maximum', df['Starting_Salary'].max())
print('Null Values', df['Starting_Salary'].isna().sum())
```

```python
df[['Starting_Salary', 'Age']].groupby('Age').mean()
```

```python
df[['Starting_Salary', 'Field_of_Study']].groupby('Field_of_Study').mean().sort_values(by='Starting_Salary')
```

I used group by and descriptive statistics like mean to help me get and understanding of what demographics on average were getting paid the most/least.

```python
df[['Starting_Salary', 'Gender']].groupby('Gender').mean()
```

```python
print('Correlation', df[['Age', 'Starting_Salary']].corr())
```

```python
print('Correlation', df[['High_School_GPA', 'University_GPA']].corr())
```

```python
print('Correlation', df[['Networking_Score', 'Job_Offers']].corr())
```

Used .corr() to understand how the differnt columns of data correlated. I was slightly surprised to see that the correlation with age and starting salary was only a 0.2.

```python
# Barchart
plt.figure(figsize=(16, 5))
sns.barplot(x='Field_of_Study', y= 'Starting_Salary', data=df)
plt.title('Starting salary of the field of study')
plt.xlabel('Field of Study')
plt.ylabel('Starting Salary')
```

```python
# Scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='SAT_Score', y= 'High_School_GPA', data=df, hue = 'Current_Job_Level')
plt.title('Correlation between SAT score and High School GPA')
plt.xlabel('SAT Score')
plt.ylabel('High School GPA')
```

```python
# Scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Networking_Score', y= 'Job_Offers', data=df, hue = 'Gender')
plt.title('Correlation between Networking score and Job Offers')
plt.xlabel('Networking Score')
plt.ylabel('Job Offers')
```

```python
# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['Career_Satisfaction'])
plt.title('Count of Self-reported satisfaction with first job (1 being very dissatisfied, 10 being very satisfied)')
plt.xlabel('Career_Satisfaction')
```
